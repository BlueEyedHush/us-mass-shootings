```python
from folium import plugins

map = folium.Map(location=[40, -100], zoom_start=4.4)
for index, row in json.iterrows():
    folium.CircleMarker([row.lat, row.long], radius=2, fill=True, popup="").add_to(map)
map
```

```python
#mass 
mass = json.query('no_k>4')
len(json.query('no_k>4'))
```

```python
mass_map = folium.Map(location=[40, -100], zoom_start=4.4)
for index, row in mass.iterrows():
    folium.CircleMarker([row.lat, row.long], radius=2, fill=True, popup="").add_to(mass_map)
mass_map
```

```python
%matplotlib inline
groupedByYear = json.date.map(lambda x: x.year)
groupedByYear.value_counts().plot.bar(figsize=(18,10))
```

```python
groupedByMonth = json.date.map(lambda x: x.month)
groupedByMonth.value_counts().sort_index().plot.bar(figsize=(18,10))
```

```python
#number of incidents by months in each year

import matplotlib.pyplot as plt 
import numpy as np

fig, ax = plt.subplots()
width = 0.15  

subplotsWithYear = []
for year in range(2014, 2017+1):
    index = year - 2014
    incidentsByMonth = json[json.date.dt.year == year].date.map(lambda x: x.month).value_counts().sort_index().values
    x = np.arange(1, len(incidentsByMonth)+1)
    subplot = ax.bar(x+index*width, incidentsByMonth, width)
    subplotsWithYear.append((year, subplot))
ax.legend(dict(subplotsWithYear).values(), dict(subplotsWithYear).keys())
plt.show()
```
