```python
#number of incidents by months in each year

import matplotlib.pyplot as plt 
import numpy as np

fig, ax = plt.subplots()
width = 0.15  

subplotsWithYear = []
for year in range(2014, 2017+1):
    index = year - 2014
    incidentsByMonth = json[json.date.dt.year == year].date.map(lambda x: x.month).value_counts().sort_index().values
    x = np.arange(1, len(incidentsByMonth)+1)
    subplot = ax.bar(x+index*width, incidentsByMonth, width)
    subplotsWithYear.append((year, subplot))
ax.legend(dict(subplotsWithYear).values(), dict(subplotsWithYear).keys())
plt.xlabel('Month')
plt.ylabel('Number of incidents')
plt.show()
```

```python
import numpy as np
noInfoAboutGuns = np.array(filter(lambda x: len(x) == 0, json.guns.values))
print('number of accidents with no info guns: ' + str(len(noInfoAboutGuns)))

gunsByAccident = np.array(filter(lambda x: len(x) != 0, json.guns.values))
gunsTypes = []
for guns in gunsByAccident:
    for gun in guns:
        gunsTypes.append(gun['type'])

unknownGunsTypes = filter(lambda x: x == 'Unknown', gunsTypes)
print('number of accidents with unknown guns: ' + str(len(unknownGunsTypes)))
gunsTypes = filter(lambda x: x != 'Unknown', gunsTypes)

uniqueTypes = list(set(gunsTypes))

from collections import Counter
print('number of available guns: ' + str(len(gunsTypes)))
counted = Counter(gunsTypes)
sortedGunTypes = sorted(counted.items(), key=lambda x: x[1])

gunTypeKey = list(map(lambda x: x[0], sortedGunTypes))
gunTypeValues = list(map(lambda x: x[1], sortedGunTypes))
gunTypeValues

fig, ax = plt.subplots()
subplot = ax.bar(range(1, len(gunTypeValues)+1), gunTypeValues, width)
ax.set_xticks(range(1, len(gunTypeValues)+1))
ax.set_xticklabels(gunTypeKey, rotation=90)
plt.xlabel('Gun type')
plt.ylabel('Number of incidents')
plt.show()
```

```python
numberGunsInAccident = map(lambda x: len(x), gunsByAccident)
countedNumberGunsInAccident= Counter(numberGunsInAccident)
sortedNumberGunsInAccident = sorted(countedNumberGunsInAccident.items(), key=lambda x: x[0])
print(countedNumberGunsInAccident)

numberOfGunsKey = list(map(lambda x: x[0], sortedNumberGunsInAccident))
numberOfGunsValues = list(map(lambda x: x[1], sortedNumberGunsInAccident))

fig, ax = plt.subplots()
subplot = ax.bar(range(1, len(numberOfGunsKey)+1), numberOfGunsValues, width)
ax.set_xticks(range(1, len(numberOfGunsKey)+1))
ax.set_ylabel('Number of incidents')
ax.set_xlabel('Number of guns')
# ax.set_xticklabels(gunTypeKey, rotation=90)
plt.show()
```

```python
groupedByState = json.groupby(['state'])
countedGroupedByState = groupedByState.count()[['city']]
sumGroupedByState = groupedByState.sum()[['no_i', 'no_k']]

concatedGroupedByState = pd.concat([countedGroupedByState, sumGroupedByState], axis=1)
concatedGroupedByState.drop(["city"], inplace=True, axis=1)
concatedGroupedByState.rename(columns = {'no_i':'Number of injured people'}, inplace=True)
concatedGroupedByState.rename(columns = {'no_k':'Number of killed people'}, inplace=True)
concatedGroupedByState

plot = concatedGroupedByState.plot.bar(figsize=(18,10))
plot.set_xlabel("State")

# Number of accidents by state
```

```python
suspects = []

for participants in json.participants:
    for participant in participants:
        if participant['type'] == 'Subject-Suspect':
            suspects.append(participant)
ages = map(lambda x: x['age'], suspects)
ages = filter(lambda x: x != None, ages)
print("Number of suspects with known age: " + str(len(ages)))

agesDataFrame = pd.DataFrame(ages, columns=['age'])
plot = agesDataFrame.age.value_counts().sort_index().plot.bar(figsize=(18,5))
plot.set_ylabel('Number of people')
plot.set_xlabel('Suspect age')
```

```python
gender = map(lambda x: x['gender'], suspects)
gender = filter(lambda x: x != None, gender)

genderDataFrame = pd.DataFrame(gender, columns=['gender'])
plot = genderDataFrame.gender.value_counts().sort_index().plot.bar(figsize=(18,5))
plot.set_ylabel('Number of people')
plot.set_xlabel('Gender')
```

```python
rowsWithGuns = json[json.apply(lambda x: len(x['guns']) > 0, axis=1)][['state', 'guns']]
rowsWithGuns

gunStateDict = {}
for index, row in rowsWithGuns.iterrows():
    for gun in row.guns:
        if gun["type"] != "Unknown":
            if row["state"] not in gunStateDict:
                gunStateDict[row["state"]] = {gun["type"]: 1}
            elif gun["type"] not in gunStateDict[row["state"]]:
                gunStateDict[row["state"]][gun["type"]] = 1
            else:
                gunStateDict[row["state"]][gun["type"]] = gunStateDict[row["state"]][gun["type"]] + 1
# print gunStateDict

for state in gunStateDict:
    print state
    values = gunStateDict[state]
    print values
```
