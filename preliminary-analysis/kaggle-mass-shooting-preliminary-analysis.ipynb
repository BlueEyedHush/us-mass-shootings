Now let's try to deal with ambiguity within 'Gender' field

```python
pd.options.display.max_colwidth = 20
ms_gender = ms2.copy()

# 'M', 'Unknown', 'Male', 'M/F', 'Male/Female', 'Female'
def gender_mapper(g):
    ng = g.lower()
    if ng == 'unknown':
        return np.nan
    elif ng == 'm':
        return 'male'
    elif ng == 'm/f':
        return 'male/female'
    else:
        return ng

ms_gender['gender_dedup'] = ms_gender['Gender'].map(gender_mapper)
ms_gender
```

```python
ms_gender['gender_dedup'].value_counts()
```

Now to mental health issues

```python
ms_missues = ms_gender.copy()

# 'Unclear', 'Yes', 'Unknown', 'No', 'unknown'
def mhi_mapper(mhi):
    mhi = mhi.lower()
    if mhi in ['unclear', 'unknown']:
        return np.nan
    else:
        return mhi
    

ms_missues['missues_dedup'] = ms_missues['Mental Health Issues'].map(mhi_mapper)
ms_missues
```

```python
ms_missues['missues_dedup'].isnull().sum()
```

```python
ms_missues['missues_dedup'].value_counts()
```

Races:

```python
ms_races = ms_missues.copy()

# 'White', 'Asian', nan, 'Black', 'Latino', 'Other', 'Unknown',
#       'Black American or African American',
#       'White American or European American', 'Asian American',
#       'Some other race', 'Two or more races',
#       'Black American or African American/Unknown',
#       'White American or European American/Some other Race',
#       'Native American or Alaska Native', 'white', 'black',
#       'Asian American/Some other race'
#
def race_mapper(mhi):
    if isinstance(mhi, float) and np.isnan(mhi):
        return mhi
    
    races = []
    mhi = mhi.lower()
    if 'unknown' in mhi:
        races.append('u')
    if 'black' in mhi or 'african' in mhi:
        races.append('black')
    if 'white' in mhi or 'european' in mhi:
        races.append('white')
    if 'asian' in mhi:
        races.append('asian')
    if 'other' in mhi:
        races.append('other')
    if 'latino' in mhi:
        races.append('latino')
    if 'two or more' in mhi or '/' in mhi:
        races.append('>1')
    if 'native' in mhi:
        races.append('native')
    
    if races == ['u']:
        return np.nan
    elif not races:
        return '!' + mhi
    else:
        return '; '.join(races)
    
ms_races['races_dedup'] = ms_races['Race'].map(race_mapper)
ms_races
```

```python
ms_races['races_dedup'].isnull().sum()
```

```python
ms_races['races_dedup'].value_counts()
```

It seems last component is always a state.
In one case we don't have city, but a state only (Washington D.C.)

```python
ms_states = ms_races.copy()
ms_states['state'] = ms_states['Location'].str.split(',').map(lambda l: l[-1].strip(), 'ignore')
ms_states
```

```python
ms_states['state'].value_counts()
```

It seems a couple of states use shortcut (e.g. CA) instead of full name (e.g. California)

```python
ms_states[ms_states['state'].str.len() <= 2]
```

```python
ms_sf = ms_states.copy()

def state_mapper(s):
    mapping = {'nv': 'nevada', 'ca': 'california', 'pa': 'pensylvania', 'wa': 'washington', 'la': 'louisiana'}
    s = s.lower()
    if s in mapping:
        return mapping[s]
    else:
        return s

ms_sf['state_full'] = ms_sf['state'].map(state_mapper, na_action='ignore')
ms_sf
```

```python
ms_sf['state_full'].value_counts()
```

```python
ms_sf['state_full'].value_counts().count()
```

## Dates

```python
split_date = ms_cfilter['Date'].str.extract('^(?P<day>[0-9]+)/(?P<month>[0-9]+)/(?P<year>[0-9]+)$')
split_date
```

```python
ms_ds = pd.concat([ms_cfilter, split_date], axis=1)
del ms_ds['Date']
ms_ds
```

```python
date_bins = pd.cut(ms_ds['year'].map(lambda x: int(x)), 17)
date_bins
```

```python
ms_dbin = ms_ds.copy()
ms_dbin['year_bin'] = date_bins
ms_dbin_cnts = ms_dbin['year_bin'].value_counts()
ms_dbin_cnts
```
